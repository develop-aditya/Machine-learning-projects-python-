```python
#importing essential lib
import matplotlib.pyplot as plt
from matplotlib import style #used to style graph
import pandas as pd
style.use("ggplot")#Grammar of Graphics"
```

```python
x=[1,2,3,4]
y=[4,5,1,1]
plt.plot(x,y)
plt.title("Info")#used to add title to whole graph
plt.xlabel("x-axis")#used to add label to x and y axis
plt.ylabel("y-axis")
plt.show()
```

```python
#ex-2
x1=[5,8,10]
y1=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.plot(x1,y1,'g',label="line one",linewidth=5)
plt.plot(x2,y2,'r',label="line two",linewidth=5)
plt.legend()#The legend helps identify the different elements within the plot, such as lines, scatter points, or bars, by associating them with labels.
plt.title("EPIC INFO")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True,color="k")# color="k" used for black color
plt.show()
```

```python
# Creation of bar Graph
days1=[0.25,1.25,2.25,3.25,4.25]
km1=[50,40,70,80,20]
days2=[0.75,1.75,2.75,3.75,4.75]
km2=[80,20,20,50,60]
plt.bar(days1,km1,label="BMW",width=0.5)#bar is used to plot bar graph
plt.bar(days2,km2,label="Audi",width=0.5)
plt.legend()
plt.title('Information')
plt.xlabel('Days')
plt.ylabel('Distance(Kms)')
plt.show()
```
